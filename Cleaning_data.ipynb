```python
import pandas as pd

# Cargar el archivo Excel en un DataFrame
ruta_archivo = '//content/TODAS_U.xlsx'  # Reemplaza con la ruta correcta
df = pd.read_excel(ruta_archivo)

df
```

```python
valores_unicos = df['INST_NOMBRE_INSTITUCION'].unique()
valores_unicos
```

```python
# Contar el número de filas con valores vacíos (nulos)
filas_vacias = df.isnull().sum(axis=1).value_counts().loc[lambda x: x.index > 0].sum()
print(f"Número de filas con valores vacíos: {filas_vacias}")

# Contar el número de columnas con valores vacíos (nulos)
columnas_vacias = df.isnull().sum(axis=0).value_counts().loc[lambda x: x.index > 0].sum()
print(f"Número de columnas con valores vacíos: {columnas_vacias}")
```

```python
df = df.drop(columns=['ESTU_CODDANE_COLE_TERMINO', 'ESTU_COD_COLE_MCPIO_TERMINO','ESTU_COLE_TERMINO'])
```

```python
import pandas as pd
from tabulate import tabulate

# Contar el número de valores vacíos (nulos) por columna
vacios_por_columna = df.isnull().sum()

# Crear un DataFrame con los resultados
tabla_vacios = pd.DataFrame({
    'Columna': vacios_por_columna.index,
    'Cantidad de Vacíos': vacios_por_columna.values
})

# Imprimir la tabla en un formato bonito usando tabulate
print(tabulate(tabla_vacios, headers='keys', tablefmt='pretty'))
```
